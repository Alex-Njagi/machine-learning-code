```python
import matplotlib.pyplot as plt
import numpy as np
# Generate a sequence of numbers from -10 to 10 with 100 steps inbetween
x = np.linspace(-10, 10, 100)
# Create a second array using sine
y = np.sin(x)
z = np.cos(x)
plt.plot(x, y, marker="x", label = "sin")
plt.plot(x, z, marker="o", label = "cosin")
plt.grid()
plt.legend()
```

```python
normal_samples = np.random.normal(0, 1, 100)
uniform_samples = np.random.uniform(0, 1, 100)

# Creating subplots
fig, axes = plt.subplots(1, 2, figsize=(12, 8))

# Plotting histograms
axes[0].hist(normal_samples, bins=30, color='blue', density=True, alpha=0.6)
axes[0].set_title("Normal distribution")
axes[1].hist(uniform_samples, bins=30, color='green', density=True, alpha = 0.6)
axes[1].set_title("Uniform distribution")
```

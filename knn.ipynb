```python
import mglearn
import numpy as np
import matplotlib.pyplot as plt
X, y = mglearn.datasets.make_forge()

np.c_[X, y]  # concatenate X and y along the second axis
# X
# y
```

```python
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.legend(["Class 0", "Class 1"], loc = 4)
plt.xlabel("X")
plt.ylabel("y")
plt.title("Scatter plot")

# Instance vs. Model Based Learning
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)
print(f"Size of training set: {X_train.shape}")
print(f"Size of testing set: {X_test.shape}")
```

```python
# Make the classifier
clf = KNeighborsClassifier(n_neighbors = 3)
clf.fit(X_train, y_train)
```

```python
# Test model's accuracy and make predictions
print("Training set accuracy {:.4f}".format(clf.score(X_train, y_train)))
print("Testing set accuracy {:.4f}".format(clf.score(X_test, y_test)))

print("Actual labels of test set: " , y_test)
print("Predicted labels of test set: " , clf.predict(X_test))

# Overfitting, underfitting, accuracy, model complexity in model graphs
```
